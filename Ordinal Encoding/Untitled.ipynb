## One hot encoding fot top categories

```python
df1.head()
```

```python
counts = df1['brand'].value_counts()
```

```python
threshold = 100
rep = counts[counts<threshold].index
```

```python
rep
```

```python
# enocoding the categores

pd.get_dummies(df1['brand'].replace(rep,'uncommon',dtype=int).sample(5)
```

```python
df2 = df1['brand'].replace(rep,'uncommon')
pd.get_dummies(df2,columns=['brand'],dtype=int).sample(5)
```
