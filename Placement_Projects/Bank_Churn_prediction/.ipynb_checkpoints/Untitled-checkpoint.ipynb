### About the features
- CustomerId: A unique identifier for each customer
- Surname: The customer's surname or last name
- CreditScore: A numerical value representing the customer's credit score
- Geography : The country where the customer resides(France,Spain or Germany)
- Gender: The customer's gender(Male or Female)
- Age: The customer's age.
- Tenure: The number of years the customer has been with the bank
- Balance: The customer's account balance
- NumOfProducts: The number of bank products the customer uses (e.g., savings account,credit card)
- HasCrCard: Whether the customer has a credit card(1 = yes,0=no)
- isActiveMember: Whether the customer is an active number(1=yes,0=no)
- EstimatedSalary: The estimated salary of the customer
- Exited: Whether the customer has churned(1 = yes,0=no)

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
#Surname didnt play any role in churn prediction
new_train = train.drop(columns=['Surname','id','CustomerId'])
categorical_train = train.select_dtypes(include='object')
numerical_train = train.select_dtypes(exclude='object')
```

```python
categorical = new_train.select_dtypes(include='object')
numerical = new_train.select_dtypes(exclude='object').drop('Exited',axis = 1)
```

```python
num_col = []
for i in list(numerical.columns):
    num_col.append(i)

num_col
```

```python
new_train
```

```python
# Step1 --> train test split
x_train,x_test,y_train,y_test = train_test_split(new_train.drop(columns=['Exited']),new_train['Exited'],test_size=0.2,random_state=12)
```

```python
x_train.head()
```

```python
# Since we do not have any missing values hence we skip imputation
```

```python
# One Hot Encoding
trf1 = ColumnTransformer([
    ('Ohe_Geography_gender',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,2])
],remainder='passthrough')
```

```python
# Scaling
trf2 = ColumnTransformer([
    ('scale',MinMaxScaler(),[0,3,4,5,6,7,8,9])
])
```

```python
trf3 = DecisionTreeClassifier()
```

```python
trf4 = RandomForestClassifier()
```

# Create pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf4',trf4)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.named_steps
```

```python
# Predict
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Cross validation using Pipeline

```python
# cross validation using cross val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, x_train,y_train,cv=5,scoring='accuracy').mean()
```

# Grid Search using Pipeline 

```python
# gridsearchcv
params = {
    'trf4__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')
grid.fit(x_train, y_train)
```
