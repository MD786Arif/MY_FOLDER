```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Define the three columns
columns = ['Fuel_Type', 'Seller_Type', 'Transmission']

#color
pastel_colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']


# Create subplots
fig, axes = plt.subplots(2,3,figsize=(15, 10))

# Loop over each column and plot a bar plot and a pie chart
for i, column in enumerate(columns):
    # Calculate the number of occurrences for each category
    category_counts = df2[column].value_counts()
    
    # Plot bar plot
    sns.countplot(data=df2, x=column, palette='pastel', ax=axes[0, i])
    axes[0, i].set_title(f"{column} Number distribution")
    
    # Plot pie chart
    axes[1, i].pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors=pastel_colors)
    axes[1, i].set_title(f"{column} Percentage distribution")

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
